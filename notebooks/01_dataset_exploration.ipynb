```python
from src.config import get_training_dataset_path

paths = get_training_dataset_path()

normal_path = paths["normal"]
abnormal_path = paths["abnormal"]
```

### Verify Dataset Structure

```python
print(normal_path.exists())
print(abnormal_path.exists())
```

The required dataset folders are available.
Current analysis will be restricted to Pump id_00 recordings.

```python
# Number of Recordings
normal_files = sorted(list(normal_path.glob("*.wav")))
abnormal_files = sorted(list(abnormal_path.glob("*.wav")))
print("Normal :", len(normal_files))
print("Abnormal :", len(abnormal_files))
```

```python
# Metadata information of the recording
sample = normal_files[0]
info = sf.info(sample)
info
```

```python
# recording duration consistency : if all recordings equally long?
durations = []
for file in normal_files:
    durations.append(sf.info(file).duration)
duration_df = pd.DataFrame(
    {"Duration (s)": durations}
    )
duration_df.describe()
```

If all recordings have approximately identical duration, windowing strategies become simpler during feature extraction.

```python
# Sample Rate Consistency : if all recordings use the same sampling rate?
sample_rates = []
for file in normal_files:
    sample_rates.append(
        sf.info(file).samplerate
    )
set(sample_rates)
```

-> All recordings share the same sampling rate

```python
# Number of channels : if all recordings are multichannel?
channels = []
for file in normal_files:
    channels.append(
        sf.info(file).channels
    )
set(channels)
```

-> Every recording contains eight synchronized microphone channels.

```python
# File Format : if all files are stored in the same audio format.
formats = []
for file in normal_files:
    formats.append(
        sf.info(file).format
    )
set(formats)
```

### Dataset Summary

```python
summary = pd.DataFrame({
    "Property":[
        "Machine",
        "Machine ID",
        "Normal files",
        "Abnormal files",
        "Channels",
        "Sample Rate"
    ],
    "Values":[
        "Pump",
        "id_00",
        len(normal_files),
        len(abnormal_files),
        8,
        sample_rates[0]
    ]
})
summary
```
