## Notebook 03 - Waveform Analysis

### Research Question
Can machine faults be distinguished directly from their time-domain waveforms?

### Objective
This notebook investigates whether visual inspection of waveforms provides meaningful information about machine condition before moving into feature extraction.

```python
# Imports
from pathlib import Path
import random

import matplotlib.pyplot as plt
from src.config import *
from src.dataset import *
from src.visualization import plot_waveforms
```

```python
dataset_path = Path("C:/MyProjects/MIMIIResearch/id_00")
normal_path = dataset_path/"normal"
abnormal_path = dataset_path/"abnormal"
```

```python
# Loading the data
normal_files = get_audio_files(normal_path)
abnormal_files = get_audio_files(abnormal_path)

print(f"Normal recordings : {len(normal_files)}")
print(f"Abnormal recordings : {len(abnormal_files)}")
```

```python
# selecting random files for plotting
random.seed(42)
normal_examples = random.sample(normal_files, 5)
abnormal_examples = random.sample(abnormal_files, 5)
```

```python
plot_waveforms(
    normal_examples,
    channel = 0,
    title = "Normal Recordings"
)
```

```python
plot_waveforms(
    abnormal_examples,
    channel = 0,
    title = "Abnormal Recordings"
)
```
